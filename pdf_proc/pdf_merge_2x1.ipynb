```python
#https://stackoverflow.com/questions/74071638/create-bookmarks-for-separate-pdf-files-using-the-filename

#directory = '.'

source_dir = os.getcwd()
directory = source_dir

pdfFileList = [f for f in listdir(source_dir) if (isfile(join(source_dir, f)) and f.endswith('.pdf'))]
pdfFileList.sort()

pathname, extension = os.path.splitext(directory)
filename = pathname.split('\\')

print('filenames = ', pdfFileList)

    #if file.endswith('.txt'):
    #    res.append(file)
```

# Get timestamp

```python
from datetime import datetime
from datetime import date

# storing the current date/time in the variable
today = date.today()
c = datetime.now()

# Displays Date/Time
time_stamp = "_%02d%02d.%s" % (today.month, today.day, c.strftime('%2H%2M%2S'))
print("time_stamp = ", time_stamp)
```

## merge pdf files

```python
merger = PdfMerger()

for pdf in pdfFileList:
    merger.append(pdf)

output_pdf = "output_pdf" + time_stamp + ".pdf"
merger.write(output_pdf)
merger.close()
```

# adding bookmarks with pdf filename

```python
# merge page by page
#output_pdf_stream = PdfWriter(open(output_pdf, "wb"))
merger = PdfMerger()
merger.append(output_pdf)

iPg=0
iFile=0
for f in pdfFileList:
  pdfFile = PdfReader(open(f, "rb"))
  print('file: %s, %d pages' % (f, len(pdfFile.pages)))
  merger.add_outline_item(title=str(pdfFileList[iFile]),page_number=iPg)
  iFile += 1
  iPg += len(pdfFile.pages)

merger.write(output_pdf)
merger.close()

```
