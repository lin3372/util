## imports

```python
import os
from os import listdir
from os.path import isfile, join
from PyPDF2 import PdfFileMerger, PdfFileReader, PdfFileWriter, PdfReader, PdfWriter , PdfMerger
from google.colab import files
import datetime
from datetime import datetime
import shutil
```

## Get timestamp

```python
from datetime import datetime
from datetime import date

# storing the current date/time in the variable
today = date.today()
c = datetime.now()

# Displays Date/Time
time_stamp = "%02d%02d.%s" % (today.month, today.day, c.strftime('%2H%2M%2S'))
print("time_stamp = ", time_stamp)
```

## get list of pdf files

```python
from google.colab import files
from tqdm.notebook import tqdm

uploaded = files.upload()
```

## Move the list of to-be-processed files to to-be-processed directory

```python
folder_orgi_files = f'orig_{time_stamp}'
os.makedirs(f'{folder_orgi_files}', exist_ok=True)

for f in tqdm(uploaded.keys(), desc="Uploading"):
    print('User uploaded file "{name}" with length {length} bytes, moving to {orig_folder}'.\
          format(name=f, length=len(uploaded[f]), orig_folder=f'{folder_orgi_files}/{f}'))
    shutil.move(f, f"{folder_orgi_files}/{f}")
```

```python
#https://stackoverflow.com/questions/74071638/create-bookmarks-for-separate-pdf-files-using-the-filename

#directory = '.'

source_dir = f'{os.getcwd()}/{folder_orgi_files}'
directory = source_dir

pdfFileList = [f for f in listdir(source_dir) if (isfile(join(source_dir, f)) and f.endswith('.pdf'))]
pdfFileList.sort()

pathname, extension = os.path.splitext(directory)
filename = pathname.split('\\')

print('source_dir =', source_dir)
print('filenames = ', pdfFileList)
```

## adding page number to files and move those files to a processed_folder

```python
# prompt: adding page number to the center bottom of pdf file 'pdf'
folder_processed_files = f'pgNum_{time_stamp}'
os.makedirs(f'{folder_processed_files}', exist_ok=True)

for pdf_filename in pdfFileList: # Changed variable name to avoid confusion
    # Create a PDF writer object for the output file
    pdf_writer = PdfWriter()

    # Open the current PDF file for reading
    try:
        with open(f'{source_dir}/{pdf_filename}', 'rb') as pdf_file:
            pdf_reader = PdfReader(pdf_file)
            print(f"Processing file: {source_dir}/{pdf_filename} ==> {folder_processed_files}/{os.path.splitext(pdf_filename)[0]}.pdf")
            print(f"\t\t\t pages={len(pdf_reader.pages)}, [w,h]=[{pdf_reader.pages[0].mediabox.width},{pdf_reader.pages[0].mediabox.height}]")

            # Iterate through the pages of the input PDF
            for i in range(len(pdf_reader.pages)):
                # Get the page
                page = pdf_reader.pages[i]

                # Create a text object for the page number
                from reportlab.pdfgen import canvas
                from reportlab.lib.units import inch
                from io import BytesIO

                buffer = BytesIO()
                c = canvas.Canvas(buffer)
                c.setFillColorRGB(0,0.6,0) #choose your font colour // green
                c.setFillAlpha(0.33)
                # Set the font and size for the page number
                c.setFont('Courier', 8)
                # Add the page number to the bottom right corner
                page_num_text = f'P{i + 1}/{len(pdf_reader.pages)} - ({pdf_filename})'
                text_width = c.stringWidth(page_num_text, 'Courier', 8)
                # Adjust position based on page dimensions
                page_width = page.mediabox.width
                page_height = page.mediabox.height
                x_pos = (float(page_width) - text_width - 0.2 * inch)/2
                y_pos = 0.05 * inch
                c.drawString(x_pos, y_pos, page_num_text)
                c.save()

                # Create a new PDF from the text object
                page_num_pdf = PdfReader(buffer)
                page_num_page = page_num_pdf.pages[0]

                # Merge the page number with the original page
                page.merge_page(page_num_page)

                # Add the modified page to the writer
                pdf_writer.add_page(page)

            # Write the output PDF to a file
            # Changed output filename to include the original filename to avoid overwriting
            output_filename = f"{folder_processed_files}/{os.path.splitext(pdf_filename)[0]}.pdf"
            with open(output_filename, 'wb') as out_file:
                pdf_writer.write(out_file)

            print(f"PDF with page numbers created: {output_filename}")
            print(' ======= ======= ======= ======= ======= ======= ')

    except Exception as e:
        print(f"Error processing file {pdf_filename}: {e}")
```

## Zip the processed folder

```python
import shutil

zip_filename = f'pgNum_{time_stamp}'
shutil.make_archive(zip_filename, 'zip', folder_processed_files)
```
