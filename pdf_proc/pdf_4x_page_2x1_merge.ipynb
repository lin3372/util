## pdf - Adding empty pages to end of the pdf
pdf_add_empty_pages_to_pdf(inPdf, outPdf, nPages)

```python
from PyPDF2 import PdfReader, PdfWriter

def pdf_add_empty_pages_to_pdf(inPdf, outPdf, nPages):
    # Read the existing PDF
    reader = PdfReader(inPdf)
    print('file: %s, %d pages' % (inPdf, len(reader.pages)))
    writer = PdfWriter()

    # Add all pages from the existing PDF
    for page in reader.pages:
        writer.add_page(page)

    # Add an empty page
    for i in range(nPages):
        writer.add_blank_page()
    #writer.add_blank_page()

    # Save the new PDF
    with open(outPdf, "wb") as output_file:
        writer.write(output_file)

# Example usage
#input_pdf_path = "input.pdf"  # Replace with your input PDF file
#output_pdf_path = "output.pdf"  # Replace with your desired output PDF file
#add_empty_page_to_pdf(input_pdf_path, output_pdf_path)
```

```python
def pdf_merge_2pg_to_1pg_inImg(input_pdf, output_pdf):
    """Merges two consecutive pages of a PDF into one.

    Args:
        input_pdf: Path to the input PDF file.
        output_pdf: Path to the output PDF file.
    """
    reader = PdfReader(input_pdf)
    writer = PdfWriter()

    num_pages = len(reader.pages)

    # Open PDF document using pypdfium2
    pdf = pdfium.PdfDocument(input_pdf)

    # Process two pages at a time
    for i in range(0, num_pages, 2):
        packet = BytesIO()
        can = canvas.Canvas(packet, pagesize=letter)
        width, height = letter  # Portrait mode

        # First page
        page1 = pdf.get_page(i)  # Get page using pypdfium2
        pil_image1 = page1.render().to_pil() # Render to PIL Image
        can.drawInlineImage(pil_image1, 0, height / 2, width, height / 2)

        # Second page (if exists)
        if i + 1 < num_pages:
            page2 = pdf.get_page(i + 1)  # Get page using pypdfium2
            pil_image2 = page2.render().to_pil() # Render to PIL Image
            can.drawInlineImage(pil_image2, 0, 0, width, height / 2)

        can.save()

        # Merge the new page
        packet.seek(0)
        merged_page = PdfReader(packet).pages[0]
        writer.add_page(merged_page)

    # Close the PDF document
    pdf.close()

    # Save the modified PDF
    with open(output_pdf, "wb") as output_file:
        writer.write(output_file)

## Example usage:
#input_pdf_path = "your_input_pdf.pdf"  # Replace with your input PDF file
#output_pdf_path = "merged_output.pdf"  # Replace with desired output path
#merge_two_pages_to_one(input_pdf_path, output_pdf_path)
```

```python
from PyPDF2 import PdfReader, PdfWriter, PageMerge

def merge_pdf_pages(input_file='input.pdf'):
    # Read the input PDF file
    reader = PdfReader(input_file)
    writer = PdfWriter()
    pages = PageMerge()
    for page in reader.pages:
        pages.add(page)
    writer.addpage(pages.render())
    with open("output.pdf",'wb') as file:
        writer.write(file)
# Example usage
#merge_pdf_pages('input.pdf')
```
